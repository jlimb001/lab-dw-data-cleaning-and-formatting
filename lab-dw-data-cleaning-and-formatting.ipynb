## Exercise 1: Cleaning Column Names

```python
# Installing panda and imprting the data

import pandas as pd
import numpy as np

url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
df = pd.read_csv(url)
df
```

## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
# Checking the uniques values 
unique_values = df['gender'].unique()
print('Unique values in education column:' , unique_values)

non_unique_values = df['gender'].nunique()
print('Number of unique values:' , non_unique_values)

value_counts = df['gender'].value_counts()
print('Frequency of unique values:' , value_counts)
```

```python
# Cleaning invalid values

# Formatting the gender column
gender_map = {
    'M': 'M',
    'F': 'F',
    'Femal': 'F',
    'Male': 'M',
    'female': 'F'
}

df['gender'] = df['gender'].map(gender_map)
df['gender']
```

```python
# Checking the uniques values 
unique_values = df['state'].unique()
print('Unique values in education column:' , unique_values)

non_unique_values = df['state'].nunique()
print('Number of unique values:' , non_unique_values)

value_counts = df['state'].value_counts()
print('Frequency of unique values:' , value_counts)
```

```python
# Formatting states into abbreviation
df['state']
state_map = {
    'Arizona': 'AZ',
    'Nevada': 'N',
    'California': 'Cali',
    'Washington': 'W'
    
}

df['state'] = df['state'].map(state_map)
df['state']
```

```python
# Checking unique values in education column
unique_values = df['education'].unique()
print('Unique values in education column:' , unique_values)

non_unique_values = df['education'].nunique()
print('Number of unique values:' , non_unique_values)

value_counts = df['education'].value_counts()
print('Frequency of unique values:' , value_counts)
```

```python
#  *In Customer Lifetime Value, delete the `%` character*

# converting all values to strings first
df['customer_lifetime_value'] = df['customer_lifetime_value'].astype(str)

# removing the % and convert back to numeric
df['customer_lifetime_value'] = df['customer_lifetime_value'].apply(lambda x: x.replace('%', '') if isinstance (x, str) else x)

# converting to numeric
df['customer_lifetime_value'] = pd.to_numeric(df['customer_lifetime_value'], errors='coerce')
df['customer_lifetime_value']
```

```python
# Checking unique values in vehicle column
unique_values = df['vehicle_class'].unique()
print('Unique values in education column:' , unique_values)

non_unique_values = df['vehicle_class'].nunique()
print('Number of unique values:' , non_unique_values)

value_counts = df['vehicle_class'].value_counts()
print('Frequency of unique values:' , value_counts)

```

```python
# - *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

# Formatting vehicle class 
# defining mapping dict
vehicle_class_map = {
    'Luxury SUV': 'Luxury',
    'Sports Car': 'Luxury',
    'Luxury car': 'Luxury'
    }

df['vehicle_class'] = df['vehicle_class'].map(vehicle_class_map)
df['vehicle_class']
```

```python
# Checking unique values in vehicle class
unique_values = df['vehicle_class'].unique()
print('Unique values in education column:' , unique_values)

non_unique_values = df['vehicle_class'].nunique()
print('Number of unique values:' , non_unique_values)

value_counts = df['vehicle_class'].value_counts()
print('Frequency of unique values:' , value_counts)
```

```python
# categorical value: customer, state, gender, education, policy type, vehicle class
# Hints for categorical : can use mode 

# Filling the missing value with mode (most frequent value) for categorical columns
df['customer'] = df['customer'].fillna('Unknown')
df['state'] = df['state'].fillna(df['state'].mode()[0]) 
df['gender'] = df['gender'].fillna(df['gender'].mode()[0]) 
df['education'] = df['education'].fillna('Not Specified') # Using a placeholder for missing education
df['policy_type'] = df['policy_type'].fillna(df['policy_type'].mode()[0]) 
df['vehicle_class'] = df['vehicle_class'].fillna('Other')

df.head(10)
```

```python
# Numerical data: customer lifetime value, income, monthly premium auto , no. of open complaints, total claim 
# The above mentioned columns likely represent continuous or discrete numerical data
# Hints for numerical: can use mean or median 

# Filling the missing value with mode (most frequent value) for categorical columns

# 1. Lifetime value: Most likely to be continuous variable representing customer value over time. Therefore, median will be used
df['customer_lifetime_value'] = df['customer_lifetime_value'].fillna(df['customer_lifetime_value'].median())
```

```python
# 2. Income: It is a continuous variable that may have a wide range of values.
df['income'] = df['income'].fillna(df['income'].median())
```

```python
# 3. Monthly premium auto: Continuous variable representing monthly premiums for auto insurance
df['monthly_premium_auto'] = df['monthly_premium_auto'].fillna(df['monthly_premium_auto'].median())
```

```python
# 4. No. of open complaints: Likely a discrete representing counts of complaints.
df['number_of_open_complaints'] = df['number_of_open_complaints'].fillna(0)
```

```python
# 5. Total claim: Likely to be a continuous variable representing the total claim amount made by the customer
df['total_claim_amount'] = df['total_claim_amount'].fillna(df['total_claim_amount'].median())

```
